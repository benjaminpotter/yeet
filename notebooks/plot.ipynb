```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
filepath = "../data/24-10-31-22-03-32.dat"
data = np.genfromtxt(filepath, delimiter=",")
print(data.shape)
```

```python
for i, row in enumerate(data):
    if i % 10 != 0:
        continue
        
    plt.plot(range(data.shape[1]), (row * 15.0) + i, color='black')

plt.title("Evolution of E-field with Time\n{}".format(filepath.split("/")[-1].split(".")[0]))
plt.xlabel("Position")
plt.ylabel("Step")
plt.savefig("./{}.svg"
plt.show()
```
